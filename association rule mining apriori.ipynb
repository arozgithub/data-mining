```python
import pandas as pd
data = pd.read_csv("/content/DATA.csv - DATA.csv.csv")

# Keep only the first 500 rows of the dataset
data = data.head(500)
data_processed = pd.get_dummies(data)
data_processed
# Binarize the dataset (convert non-zero values to 1)
data_processed[data_processed != 0] = 1
data_processed
```

```python
#a

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import pandas as pd

# Perform one-hot encoding and find frequent itemsets
frequent_itemsets = apriori(data_processed, min_support=0.1, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

# Rank top 10 association rules for different combinations of support and confidence
support_confidence_combinations = [(0.1, 0.7), (0.2, 0.8), (0.3, 0.9)]

for support, confidence in support_confidence_combinations:
    print(f"\nTop 10 association rules with support={support} and confidence={confidence}:\n")
    filtered_rules = rules[(rules['support'] >= support) & (rules['confidence'] >= confidence)]
    top_10_rules = filtered_rules.sort_values(by='lift', ascending=False).head(10)
    print(top_10_rules)
```

```python
#b

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import pandas as pd

# Perform one-hot encoding and find frequent itemsets
frequent_itemsets = apriori(data_processed, min_support=0.1, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

# Define a function to calculate interest
def calculate_interest(rule):
    return rule.confidence - rule.support / len(data_processed)

# Calculate interest for each rule
rules['interest'] = rules.apply(calculate_interest, axis=1)

# Rank top 10 association rules for different combinations of support and interest
support_interest_combinations = [(0.1, 0.2), (0.2, 0.3), (0.3, 0.4)]

for support, interest in support_interest_combinations:
    print(f"\nTop 10 association rules with support={support} and interest={interest}:\n")
    filtered_rules = rules[(rules['support'] >= support) & (rules['interest'] >= interest)]
    top_10_rules = filtered_rules.sort_values(by='lift', ascending=False).head(10)
    print(top_10_rules)
```

```python
#c
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules


# Select relevant attributes
selected_attributes = ['age', 'education', 'occupation', 'workclass', 'race', 'sex', 'marital-status', 'native-country']

# Subset the data with selected attributes
data_subset = data[selected_attributes]

# Convert categorical variables into binary dummy variables
data_processed = pd.get_dummies(data_subset)

# Binarize the dataset (convert non-zero values to 1)
data_processed[data_processed != 0] = 1

# Perform one-hot encoding and find frequent itemsets
frequent_itemsets = apriori(data_processed, min_support=0.1, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

# Show at least 10 rules
print("Top 10 association rules:\n")
print(rules.head(10))
```
